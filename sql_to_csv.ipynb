```python
data  =pd.read_csv('all_features.csv')
```

```python
data.dtypes
```

```python
data['sp_Date'] = pd.to_datetime(data['sp_Date'])
```

```python
data.dtypes
```

```python
date_index = pd.DatetimeIndex(data['sp_Date'])

# 提取年份
year = date_index.year

# 根据条件筛选数据
filtered_data = data[year >= 2021]
```

```python
filtered_data
```

```python
filtered_data.isnull().sum()
```

```python
missing_values = filtered_data.isnull().sum()
for column, count in missing_values.items():
    if count > 0:
        print(f"{column}\t\t{count}")
```

```python
df = filtered_data.copy()
```

```python
df['Month'] = df['sp_Date'].dt.month
df['Year'] = df['sp_Date'].dt.year


grouped = df.groupby(['sp_Symbol', 'Month','Year'])

df['sp_Sentiment_scores'] = grouped['sp_Sentiment_scores'].transform(lambda x: x.fillna(x.mean()))
df['sp_ESG_score'] = grouped['sp_ESG_score'].transform(lambda x: x.fillna(x.mean()))
df['sp_E_score'] = grouped['sp_E_score'].transform(lambda x: x.fillna(x.mean()))
df['sp_S_score'] = grouped['sp_S_score'].transform(lambda x: x.fillna(x.mean()))
df['sp_G_score'] = grouped['sp_G_score'].transform(lambda x: x.fillna(x.mean()))
df['sp_search_popularity'] = grouped['sp_search_popularity'].transform(lambda x: x.fillna(x.mean()))


# df['sp_Sentiment_scores'].fillna(0, inplace=True)
```

```python
missing_values = df.isnull().sum()
for column, count in missing_values.items():
    if count > 0:
        print(f"{column}\t\t{count}")
```

```python
df.to_csv('all_features_filled.csv')
```
