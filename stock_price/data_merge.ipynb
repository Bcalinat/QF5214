```python
import pandas as pd
```

```python
Comps = ["MSFT", "AAPL", "NVDA", "AVGO", 
         "BRK-B", "JPM",
         "LLY", "UNH", "AMZN", "GE", "GOOG", "WMT", "XOM", "LIN",]
```

```python
t = []
for i in range(len(Comps)):
    t.append(pd.read_csv(f'../data/{Comps[i]}_price.csv'))
    t[i]['Symbol']=Comps[i]
```

```python
df = t[0]
for i in range(1,len(Comps)):
    df = pd.concat([df,t[i]],axis = 0)
df
```

```python
df.to_csv('../data/stock_price.csv')
```
