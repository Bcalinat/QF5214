```python
import pandas as pd
```

```python
search_popularity = pd.read_csv('../data/search_popularity.csv')
stock_price = pd.read_csv('../data/stock_price.csv')
yahoo_esg_score = pd.read_csv('../data/Yahoo_Esg_Score.csv')
sentiment_scores_daily = pd.read_csv('../data/sentiment_scores/sentiment_scores_daily.csv')
```

```python
search_popularity
```

```python
# stock_price.iloc[:,1:]
stock_price = stock_price.iloc[:,1:]
```

```python
search_popularity_melt = search_popularity.melt(id_vars=['Date'],
                                                value_vars=["MSFT", "AAPL", "NVDA", "AVGO", 
                                                                "BRK-B", "JPM",
                                                                "LLY", "UNH", "AMZN", "GE", "GOOG", "WMT", "XOM", "LIN"],
                                                var_name='Symbol',
                                                value_name='search_popularity')
```

```python
search_popularity_melt
```

```python
data = stock_price.merge(search_popularity_melt,how='outer',on = ['Date','Symbol'])
```

```python
data
```

```python
yahoo_esg_score
```

```python
yahoo_esg_score.rename(columns = {'symbol':'Symbol'},inplace=True)
```

```python
data = data.merge(yahoo_esg_score,how = 'outer',on = ['Date','Symbol'])
```

```python
data.merge(sentiment_scores_daily,how = 'outer',on = ['Symbol','Date'])
```

```python
data.sort_values(by = ['Symbol','Date'],inplace=True)
```

```python
data
```

```python
data.columns
```

```python
columns = ['Date', 'Symbol', 'Open', 'High', 'Low', 'Close', 'Adj Close', 'Volume', 
       'search_popularity', 'esg_score', 'indicator']
data = data[columns]
```

```python
data.to_csv('../data/features.csv')
```
