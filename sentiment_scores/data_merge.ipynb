```python
import pandas as pd
from datetime import datetime
```

```python
Comps_s = ["MSFT", "AAPL", "NVDA", "AVGO", 
        #  "BRK-B", "JPM",
         "LLY", "UNH", "AMZN", "GE", "GOOG", "WMT", "XOM", "LIN",]
```

```python
t = []
for i in range(len(Comps_s)):
    t.append(pd.read_csv(f'../data/sentiment_scores/{Comps_s[i]}.csv'))
    t[i]['Unnamed: 0']=Comps_s[i]
```

```python
sentiment_scores = t[0]
for i in range(1,len(Comps_s)):
    sentiment_scores = pd.concat([sentiment_scores,t[i]],axis = 0)
sentiment_scores
```

```python
sentiment_scores.rename(columns={'Unnamed: 0':'Symbol','time':'Date','scores':'Sentiment_scores'},inplace = True)
```

```python
sentiment_scores
```

```python
sentiment_scores['Date']=sentiment_scores['Date'].apply(lambda x: datetime.strptime(x[:8], "%Y%m%d").strftime("%Y-%m-%d"))
```

```python
sentiment_scores_daily = sentiment_scores.groupby(by = ['Symbol','Date'])[['Sentiment_scores']].mean().reset_index()
```

```python
sentiment_scores_daily
```

```python
sentiment_scores_daily.to_csv('../data/sentiment_scores/sentiment_scores_daily.csv')
```
